```python
df3 = pd.DataFrame({
    'Data': [
        '1987_M_US _1',
        '1990?_M_UK_1',
        '1992_F_US_2',
        '1970?_M_   IT_1',
        '1985_F_I  T_2'
]})
df3
```

```python
df3 = df3.Data.str.split('_',expand=True)
df3
```

```python
df3.columns = ['Year','Sex','Country','# children']
df3
```

```python
df3.Country = df3.Country.str.replace(' ', '')
df3
```

```python
df3.Year = df3.Year.str.replace('?','')
df3
```
