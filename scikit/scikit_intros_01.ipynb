```python
import pandas as pd

# load data
df = pd.read_csv('titanic.csv')

print(df.head())
print(df.isnull().sum()) # check missing values
print(df.info()) # investigate data types
```

```python
from sklearn.impute import SimpleImputer

numeric_features = ['Age', 'Fare'] # numeric columns

imputer = SimpleImputer(strategy='mean') # filling missing values with the means of that column
df[numeric_features] = imputer.fit_transform(df[numeric_features])
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# (Male: 0, Female: 1)
df['Sex'] = LabelEncoder().fit_transform(df['Sex']) # making sex value numeric

# convert Embarked column with onehotencoder
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

# dependent and independent values for model
X = df[['Pclass', 'Sex', 'Age', 'Fare', 'Embarked_Q', 'Embarked_S']]
y = df['Survived']

# split data as train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression

# define and train the model
model = LogisticRegression()
model.fit(X_train, y_train)

# make prediction with test set
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

# accuracy of the model
print("Accuracy:", accuracy_score(y_test, y_pred))

# Confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
